# analysis data:


### Univariate Analysis

```python
#1 How are movie ratings distributed?
sns.histplot(df['Rating'], bins=20)
plt.show()
```

```python
#2. What is the runtime distribution?
sns.histplot(df['Runtime (Minutes)'])
plt.show()
```

```python
#3. How is revenue distributed?
sns.boxplot(
x=df['Revenue (Millions)']
)
```

```python
#4 How are metascores distributed?
sns.histplot(df['Metascore'])

#output
print("Most movies score: 50–75")
```

```python
#5 Which genres are most common?
df['Main_Genre']=df['Genre'].str.split(',').str[0]
df['Main_Genre'].value_counts().head(10)

#output:
#print("Usually: Action , Drama , Comedy")
```

```python
#6 Does higher rating lead to higher revenue?
sns.scatterplot(
x='Rating',
y='Revenue (Millions)',
data=df
)

#output:
print("Positive relationship exists.")
#insight
#Highly rated movies often earn more money.
```

```python
#7 Do longer movies get better ratings?
sns.scatterplot(
x='Runtime (Minutes)',
y='Rating',
data=df
)

#output:
print("Weak relationship.")
#insight: Long runtime does not guarantee success.
```

```python
#8 Do more votes mean higher revenue?
sns.scatterplot(
x='Votes',
y='Revenue (Millions)',
data=df
)

#output:
print("Strong positive relationship")
#insight: Popular movies generate more revenue.
```

```python
#9. Which genres receive higher ratings?
sns.boxplot(
x='Main_Genre',
y='Rating',
data=df
)

plt.xticks(rotation=90)
#output:
print("Some genres consistently achieve better ratings")
```

```python
#10. Which movies have runtime ≥180 minutes?
long_movies=df[
df['Runtime (Minutes)']>=180]
long_movies[
['Title',
'Runtime (Minutes)',
'Rating']
]

#insight: Very long movies are rare.
```

```python
#11. Which year received the highest votes?
df.groupby('Year')['Votes'] \
.sum() \
.sort_values(ascending=False)

#Shows peak audience engagement years.
```

```python
#12. Which year generated the highest revenue?
df.groupby('Year')[
'Revenue (Millions)'
].sum().sort_values(
ascending=False
)

#Identifies strongest box-office years.
```

```python
#13. What is the average rating by director?
df.groupby('Director')[
'Rating'
].mean().sort_values(
ascending=False
)

#Finds most consistent directors.
```

```python
#14. How many movies were produced each year?
df['Year'].value_counts().sort_index()
sns.countplot(
x='Year',
data=df
)

plt.xticks(rotation=90)

#Shows movie production trends.
```

```python
#15. classification of movies into categories such as "Excellent," "Good," and "Average" based on ratings

def category(rating):
    if rating >= 8:
        return "Excellent"
    
    elif rating >= 7:
        return "Good"
    
    else:
        return "Average"

df['Category'] = df['Rating'].apply(category)

print(df[['Title', 'Rating', 'Category']].head())
```

```python
#counting_category
print(df['Category'].value_counts())
```

```python
df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Movie Categories')
plt.ylabel('')
plt.show()
```

```python
#16 Which variables are strongly related?
plt.figure(figsize=(10,6))

sns.heatmap(
df.corr(numeric_only=True),
annot=True
)

plt.show()

#Popular movies often earn more money.
```
