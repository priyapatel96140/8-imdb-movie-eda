```python
#1 identifying movies with exceptionally long runtimes (greater than or equal to 180 minutes)

long_movies = df[df['Runtime (Minutes)'] >= 180]
print(long_movies[['Title', 'Runtime (Minutes)', 'Year']])
```

```python
plt.figure(figsize=(8,5))
plt.bar(long_movies['Title'], long_movies['Runtime (Minutes)'])
plt.xticks(rotation=90)
plt.title('Movies with Runtime >= 180 Minutes')
plt.show()
```

```python
#2. movies perform differently in terms of ratings, revenue, or audience:
print(df[['Rating', 'Revenue (Millions)', 'Votes']].describe())
```

```python
#3. which years saw the highest number of votes?

votes_year = df.groupby('Year')['Votes'].sum()

votes_year.plot(kind='bar')
plt.title('Total Votes by Year')
plt.ylabel('Votes')
plt.show()
```

```python
#4 which periods of heightened audience engagement:
engagement = df.groupby('Year')['Votes'].mean()
engagement.plot(kind='line', marker='o')
plt.title('Audience Engagement by Year')
plt.ylabel('Average Votes')
plt.show()
```

```python
#5 Revenue Over Time
revenue_year = df.groupby('Year')['Revenue (Millions)'].sum()

revenue_year.plot(kind='bar')
plt.title("Revenue by Year")
plt.show()
```

```python
#6 Correlation Between Year and Rating
correlation = df['Year'].corr(df['Rating'])
print("Correlation:", correlation)
```

```python
#7 Revenue Trend Analysis:
avg_revenue = df.groupby('Year')['Revenue (Millions)'].mean()
print(avg_revenue)

plt.scatter(df['Year'], df['Rating'])
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Year vs Rating')
plt.show()
```

```python
#8 which years were more financially successful

success_year = df.groupby('Year')['Revenue (Millions)'].sum()
print(success_year.sort_values(ascending=False))

success_year.plot(kind='bar')
plt.title('Financial Success by Year')
plt.ylabel('Revenue')
plt.show()
```

```python
#9. Average Rating by Director:
director_rating = df.groupby('Director')['Rating'].mean()

director_rating.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top Directors by Average Rating')
plt.show()
```

```python
#10. Number of Movies Produced Each Year:
movies_year = df.groupby('Year')['Title'].count()

movies_year.plot(kind='bar')
plt.title('Movies Produced Per Year')
plt.ylabel('Number of Movies')
plt.show()
```

```python
#11. movie production is increasing or decreasing and how that might affect overall market dynamics.

movies_year = df.groupby('Year')['Title'].count()

movies_year.plot(kind='line', marker='o')
plt.title('Movie Production Trend')
plt.show()
```

```python
#12. Relationship Between Rating and Revenue
print(df[['Rating', 'Revenue (Millions)']].corr())

plt.scatter(df['Rating'], df['Revenue (Millions)'])

plt.xlabel('Rating')
plt.ylabel('Revenue')
plt.title('Rating vs Revenue')
plt.show()
```

```python
#13. do higher-rated movies tend to generate higher revenue?
high_rate=df[df['Rating']>=7]
print(high_rate['Revenue (Millions)'].mean())

sns.scatterplot(x='Rating',
                y='Revenue (Millions)',
                data=df)

plt.title('Rating vs Revenue')
plt.show()
```

```python
#14. compare with all movie:
print("Average Revenue of All Movies:")
print(df['Revenue (Millions)'].mean())

print("Average Revenue of Movies Rating >= 7:")
print(high_rate['Revenue (Millions)'].mean())
```

```python
#15. classification of movies into categories such as "Excellent," "Good," and "Average" based on ratings

def category(rating):
    if rating >= 8:
        return "Excellent"
    
    elif rating >= 7:
        return "Good"
    
    else:
        return "Average"

df['Category'] = df['Rating'].apply(category)

print(df[['Title', 'Rating', 'Category']].head())
```

```python
#counting_category
print(df['Category'].value_counts())
```

```python
df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Movie Categories')
plt.ylabel('')
plt.show()
```
